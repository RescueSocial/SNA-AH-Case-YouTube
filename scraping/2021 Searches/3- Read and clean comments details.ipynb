# Read and clean comments details

```python
import pandas as pd
import json
import requests
from io import StringIO
from csv import reader
import os
import glob
```

```python
# read josn files and make a list with json files
# (make 1 change)
folder_name = 'JSON batch4-6'
path_to_json = r"C:\Users\David\Amber Heard Case\Youtube\Data\videos comments info - final -batch4/"+ folder_name
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
json_files;  
len(json_files)
```

```python
# create a data frame with all json files:

df = pd.DataFrame()
for file in json_files:
    file_path = path_to_json+"/" + file
    with open(file_path) as data_file:    
        data = json.load(data_file)
        df = df.append(pd.json_normalize(data))
df.shape
```

```python
# save every 500 video comments to one file
save_url = r'C:\Users\David\Amber Heard Case\Youtube\Data\combined files for videos comments info batch {}\scraped comments- JSON -{} Batch-{}.csv'.format(list(folder_name)[10], folder_name.split('-')[1], list(folder_name)[10])
df.to_csv(save_url, index=None)
```
