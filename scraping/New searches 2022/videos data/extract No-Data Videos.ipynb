```python
import pandas as pd
import json
import requests
from io import StringIO
from csv import reader
import os
import glob
import numpy as np
import shutil
import time
import datetime
```

```python
path_to_scraped = r'C:\Users\Administrator\AH case\YouTube\SNA-AH-Case-YouTube\scraping\New searches 2022\videos details'
scraped_files = [pos_csv for pos_csv in os.listdir(path_to_scraped) if pos_csv.endswith('.csv')]
len(scraped_files)
```

```python
path_to_ids = r'C:\Users\Administrator\AH case\YouTube\SNA-AH-Case-YouTube\scraping\New searches 2022\searches results\cleaned'
ids_files = [pos_csv for pos_csv in os.listdir(path_to_ids) if pos_csv.endswith('.csv')]
len(ids_files)
```

```python
scraped_files[0].split('--')[0][:-1]+'.csv'
```

```python
nodata_ids = []
for file in scraped_files:
    scrabed_df = pd.read_csv(path_to_scraped+'/'+file, usecols=['id'])
    scrabed_list = list(scrabed_df['id'])
    ids_df = pd.read_csv(path_to_ids+'/'+file.split('--')[0][:-1]+'.csv')
    ids_list = list(ids_df['id'])
    no_data = set(ids_list) - set(scrabed_list)
    no_data = list(no_data)
    nodata_ids.extend(no_data)
```

```python
nodata_df = pd.DataFrame(nodata_ids, columns=['id'])
```

```python
nodata_df['url'] = 'https://www.youtube.com/watch?v='+nodata_df['id']
nodata_df
```
