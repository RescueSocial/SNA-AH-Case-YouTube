## get ids of  comments made  in 4 months

```python
comments21["snippet.topLevelComment.snippet.publishedAt"] = pd.to_datetime(
    comments21["snippet.topLevelComment.snippet.publishedAt"]
)
```

### get comments of 4 months

```python
comments4months = comments21[
    (
        comments21["snippet.topLevelComment.snippet.publishedAt"].astype(str)
        >= "2021-01-01 00:00:00"
    )
    & (
        comments21["snippet.topLevelComment.snippet.publishedAt"].astype(str)
        < "2021-05-01 00:00:00"
    )
]
```

### get ids of  4 months comments

```python
comments_channels_4_months = comments4months[
    "snippet.topLevelComment.snippet.authorChannelId.value"
].drop_duplicates().reset_index().drop(columns=["index"])
```
