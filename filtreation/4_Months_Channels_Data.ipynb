## Comments data of 2021

```python
commments_data_2021 = pd.read_csv(
    "../../[CSV] Combined by year [6893 videos](videos, comments and channels) (final 25Dec21)/comments filter by year(based on comment year Not main video)/comments_data_2021.csv",
    low_memory=False,
)
```

```python
for i in commments_data_2021.columns:
    print(i)
```

```python
commments_data_2021.shape[0]
```

```python
commments_data_2021 = commments_data_2021.drop(columns=["Unnamed: 0"])
```

```python
commments_data_2021 = commments_data_2021.drop_duplicates()
```

```python
commments_data_2021.shape[0]
```

```python
commments_data_2021["snippet.topLevelComment.snippet.publishedAt"] = pd.to_datetime(
    commments_data_2021["snippet.topLevelComment.snippet.publishedAt"]
)
```

```python
commments_data_2021_4_months = commments_data_2021[
    (
        commments_data_2021["snippet.topLevelComment.snippet.publishedAt"].astype(str)
        >= "2021-01-01 00:00:00"
    )
    & (
        commments_data_2021["snippet.topLevelComment.snippet.publishedAt"].astype(str)
        < "2021-05-01 00:00:00"
    )
]
```

```python
commments_data_2021_4_months.shape[0]
```

```python
# channels commented in 2021
commments_data_2021_4_months[
    "snippet.topLevelComment.snippet.authorChannelId.value"
].nunique(), commments_data_2021_4_months[
    "snippet.topLevelComment.snippet.authorChannelId.value"
].shape[0]
```

```python
commments_data_2021_4_months
```

```python
# channels contributed in 4 months based on comments data
channels_4_months = (
    commments_data_2021_4_months[
        "snippet.topLevelComment.snippet.authorChannelId.value"
    ]
    .drop_duplicates()
    .reset_index()
    .drop(columns=["index"])
)
```

## Doing a merge between comments
- get information of channels user who commented in the 4 months (channels_4_months) from the whole channel data of 2021.

```python
channels_4_months = pd.merge(
    channels_4_months,
    channels_data_2021,
    left_on="snippet.topLevelComment.snippet.authorChannelId.value",
    right_on=["id"],
    how="inner"
            
).drop(columns=["snippet.topLevelComment.snippet.authorChannelId.value"])
```
