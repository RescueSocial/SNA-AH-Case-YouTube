## Channels

```python
ls "../[CSV] Combined by year [6893 videos](videos, comments and channels) (final 25Dec21)/combined channels info by year/"
```

```python
channels_data_path = "../[CSV] Combined by year [6893 videos](videos, comments and channels) (final 25Dec21)/combined channels info by year/"
channels_records = []
for file in os.listdir(channels_data_path):
    data_file_path = channels_data_path + file
    temp_df = pd.read_csv(data_file_path)
    channels_records.append([file.split(".")[0].split("_")[-1], temp_df.shape[0]])
```

```python
print(tabulate(channels_records, headers=["Year", "n_records"]))
```

## Comments

```python
df_comments_21 = pd.read_csv(
    "../[CSV] Combined by year [6893 videos](videos, comments and channels) (final 25Dec21)/comments filter by year(based on comment year Not main video)/comments_data_2021.csv",
    low_memory=False,
)
```

```python
df_comments_21.head(2)
```

```python
for c in df_comments_21.columns:
    print(c)
```

```python
df_comments_21["snippet.topLevelComment.snippet.publishedAt"] = pd.to_datetime(
    df_comments_21["snippet.topLevelComment.snippet.publishedAt"]
)
```

```python
df_comments_21 = df_comments_21[
    (
        df_comments_21["snippet.topLevelComment.snippet.publishedAt"].astype(str)
        >= "2021-01-01 00:00:00"
    )
    & (
        df_comments_21["snippet.topLevelComment.snippet.publishedAt"].astype(str)
        < "2021-05-01 00:00:00"
    )
]
```

```python
df_comments_21["snippet.topLevelComment.snippet.publishedAt"].max(), df_comments_21[
    "snippet.topLevelComment.snippet.publishedAt"
].min()
```

```python
df_comments_21 = df_comments_21.reset_index().drop(columns=["index", "Unnamed: 0"])
```

```python
df_comments_21.to_csv("df_comments_21_4_months.csv", index=False)
```

```python
df_comments_21.to_json("df_comments_21_months.json", orient="records")
```
