| column  | Description  | New Name  |
|:--|:--|:--|
| etag | comment entity tag **UNIQUE** | - |
| id | comment id | - |
| snippet.videoId | The ID of the video that the comments refer to, if any. If this property is not present or does not have a value, then the thread applies to the channel and not to a specific video. | video_id |
| snippet.topLevelComment.etag | etag of the top level ccomment | tl_etag |
| snippet.topLevelComment.id | ID of the top level ccomment | tl_id |
| snippet.topLevelComment.snippet.textOriginal  | The original, raw text of the comment as it was initially posted or last updated. The original text is only returned if it is accessible to the authenticated user, which is only guaranteed if the user is the comment's author. | tl_text |
|snippet.topLevelComment.snippet.authorDisplayName | The display name of the user who posted the comment. | tl_author_dname |
|snippet.topLevelComment.snippet.authorChannelId.value | The ID of the comment author's YouTube channel, if available.| tl_author_ch_id |
| snippet.topLevelComment.snippet.likeCoun | Top Level comment Likes | tl_nlikes |
| snippet.topLevelComment.snippet.publishedAt | The date and time when the comment was orignally published. The value is specified in ISO 8601 format. | tl_p_dtime |
| snippet.topLevelComment.snippet.updatedAt | The date and time when the comment was last updated. The value is specified in ISO 8601 format. | tl_u_dtime |
| snippet.totalReplyCount | The total number of replies that have been submitted in response to the top-level comment.| tl_nreplies |
| snippet.topLevelComment.snippet.moderationStatus | The comment's moderation status. This property is only returned if the API request was authorized by the owner of the channel or the video on which the requested comments were made. In addition, note that this property is not set if the API request used the id filter parameter. Valid values for this property are: - heldForReview - likelySpam - published- rejected | tl_moderation_status
