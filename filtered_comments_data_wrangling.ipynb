```python
# Create a data frame with the number of comments in each date
df_creations = df_final.groupby(['date', 'year', 'month', 'dayofmonth', 'hour']).agg(
    {'text':'count', 'nreplies': 'sum', 'nlikes': 'sum'}).reset_index().rename(
    columns={'text':'ncomments'}).sort_values('ncomments', ascending=False)

df_creations["date"] = pd.to_datetime(df_creations["date"])
df_creations.head(1)
```

```python
df_creations.ncomments = df_creations.ncomments.astype('int')
df_creations.nreplies = df_creations.nreplies.astype('int')
df_creations.nlikes = df_creations.nlikes.astype('int')
```

```python
df_creations.sort_values('ncomments', ascending=False, inplace=True)
df_creations.head(15)
```
